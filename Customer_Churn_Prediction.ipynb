```python
#importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
dataset=pd.read_csv('/content/Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
#dropping the useless columns
dataset=dataset.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
dataset.head()
```

LABEL ENCODING

```python
dataset['Geography'].unique()
```

```python
dataset=pd.get_dummies(dataset)
```

```python
dataset.info()
```

```python
dataset[['Geography_Germany','Geography_Spain','Gender_Male','Gender_Female']]=dataset[['Geography_Germany','Geography_Spain','Gender_Male','Gender_Female']].astype(int)
```

```python
dataset.head()
```

```python
dataset
```

```python
dataset.info()
```

HANDLING IMBALANCED DATA

```python
dataset['Exited'].value_counts()
#we can see there is a huge difference between 0 and 1
```

```python
x=dataset.drop('Exited',axis=1)
y=dataset['Exited']
```

```python
y
```

Resampling imbalanced data by using SMOTE method(it evens the data samples to make it balanced)

```python
x_res,y_res=SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

Train Test Splitting

```python
x_train,x_test,y_train,y_test=train_test_split(x_res,y_res,test_size=0.2,random_state=42)
```

Feature Scaling

```python
ss=StandardScaler()
```

```python
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
x_train
```

```python
x_test
```

Using machine learning algorithm for prediction . Here Logistic Regression is used

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
```

Using Different Metrics to predict accuracy

```python
accuracy_score(y_test,y_pred1)
```

```python
#after using SMOTE
accuracy_score(y_test,y_pred1)
```

As our dataset is imbalanced we will also use metrices like precision,recall and f1 scores

```python
precision_score(y_test,y_pred1)
```

```python
#after using SMOTE
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
#after using SMOTE
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

```python
#after using SMOTE
f1_score(y_test,y_pred1)
```
