```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.utils import resample
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,f1_score
```

```python
dataset=pd.read_csv('/content/fraudTest.csv')
```

```python
dataset.head()
```

```python
df=dataset[['category','amt','gender','city','state','lat','long','job','is_fraud']]
```

```python
df
```

```python
fraud=df.loc[df['is_fraud']==1]
normal=df.loc[df['is_fraud']==0]
print(fraud.shape,normal.shape)
```

As we have huge difference in the numbers our dataset is imbalanced so we will resample it

```python
fraud_unsampled=resample(fraud,n_samples=len(normal),random_state=123)
mydataset=pd.concat([fraud_unsampled,normal])
print(fraud_unsampled.shape,normal.shape)
```

```python
mydataset['is_fraud'].value_counts()
```

```python
mydataset
```

Label encoding so that we get binary values where column values are unique

```python
mydataset[['category','gender','city','state','job']].nunique()
```

```python
le=LabelEncoder()
```

```python
le1=LabelEncoder()
cat=le1.fit(mydataset['category'])
mydataset['category_lbl']=le1.transform(mydataset['category'])

le2=LabelEncoder()
gen=le2.fit(mydataset['gender'])
mydataset['gender_lbl']=le2.transform(mydataset['gender'])

le3=LabelEncoder()
cy=le3.fit(mydataset['city'])
mydataset['city_lbl']=le3.transform(mydataset['city'])

le4=LabelEncoder()
ste=le4.fit(mydataset['state'])
mydataset['state_lbl']=le4.transform(mydataset['state'])

le5=LabelEncoder()
jb=le5.fit(mydataset['job'])
mydataset['job_lbl']=le5.transform(mydataset['job'])
```

```python
le3.inverse_transform([9])#city
```

```python
le4.inverse_transform([8]) #state
```

```python
le5.inverse_transform([27]) #job
```

```python
mydataset
```

```python
df2=mydataset.drop(['category','gender','city','state','job'],axis=1)
```

```python
#now this is the final table in which we will do our prediction
#it has numerical values only
df2.head()
```

Splitting the data

```python
x=df2.drop('is_fraud',axis=1)
y=df2['is_fraud']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=123)
```

Using ML algorithm. Here Logistic Regression will be used

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
precision_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

Example usecase

```python
input_data=(9.65,30.8985,-87.1332,0,0,363,8,217)
input_data_as_nparray=np.asarray(input_data)

#reshaping the array as we are predicting for one instance

input_data_reshaped=input_data_as_nparray.reshape(1,-1)

prediction=lr.predict(input_data_reshaped)
print(prediction)

if(prediction[0]==0):
  print('Not Fraud')
else:
  print('Fraud')
```
